## Compuerta AND

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
import matplotlib.pyplot as plt  

# Red neuronal artificial para la compuerta lgica AND

# Combinaciones binarias de la compuerta AND
training_data_and = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# Etiquetas de los resultados con compuerta AND
target_data_and = np.array([[0],[0],[0],[1]], "float32")

model = Sequential()
model.add(Dense(10, input_dim=2, activation='selu'))
model.add(Dense(1,activation='selu'))
 
model.compile(loss='mean_squared_error',optimizer='sgd', metrics=['binary_accuracy'])

 
history = model.fit(training_data_and, target_data_and, epochs=350)
 
# evaluamos el modelo
scores = model.evaluate(training_data_and, target_data_and)

#impresin de resultados
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data_and).round())

# Grfica de loss vs epochs AND
loss_train = history.history['loss']
acc = history.history['binary_accuracy']
epochs = range(1,351)
plt.plot(epochs, loss_train, 'g', label='Training loss')
plt.title('Loss AND')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Grfica de acc vs epochs AND
plt.plot(epochs, acc, 'b', label='Accuracy')
plt.title('Accuracy AND')
plt.xlabel('Epochs')
plt.ylabel('ACC')
plt.legend()
plt.show()
```
